```python
import random  ##import random library for generating random number in the game from the player opponent

print('Rock Paper Scissors')
print('-------------------')
## representing rock, paper, scissors by numbers
print('enter 0 for Rock')
print('enter 1 for Paper')
print('enter 2 for Scissors')

print('-------------------')
## storing the player and computer score in variables
player_score=0
computer_score=0

##while loop for controlling the end of the game by asking the user if he wants to continue
while True:
    pl_turn=int(input('Player turn ')) ## storing the player turn in a variable

    c_turn=random.randint(0,2) ##input from computer by specifying the inputs between 0 and 2
    ## printing player turn
    if pl_turn==0:
        print('Player turn is Rock ')
    elif pl_turn==1:
        print('Player turn is Paper ')
    elif pl_turn==2:
        print('Player turn is Scissors ')
    else:
        print('undefind turn')
    
    ## printing computer turn
    if c_turn==0:
        print('Computer turn is Rock ')
    elif c_turn==1:
        print('Computer turn is Paper ')
    elif c_turn==2:
        print('Computer turn is Scissors ')
    else:
        print('Undefind turn')
        
    ## specifying the conditions, incremnting the results
    if pl_turn==0 and c_turn==0:
        print('Draw')
        print('Try again')
        print('-------------------')
        continue
    elif pl_turn==0 and c_turn==1:
        print('Computer wins')
        computer_score+=1
    elif pl_turn==0 and c_turn==2:
        print('Player wins')
        player_score+=1
    elif pl_turn==1 and c_turn==0:
        print('Player wins')
        player_score+=1
    elif pl_turn==1 and c_turn==1:
        print('Draw')
        print('Try again')
        print('-------------------')
        continue
    elif pl_turn==1 and c_turn==2:
        print('Computer wins')
        computer_score+=1
    elif pl_turn==2 and c_turn==0:
        print('Computer wins')
        computer_score+=1
    elif pl_turn==2 and c_turn==1:
        print('Player wins')
        player_score+=1
    elif pl_turn==2 and c_turn==2:
        print('Draw')
        print('Try again')
        print('-------------------')
        continue
    
    print('-------------------------------')
    print('Result : ',player_score,' - ',computer_score)
    print('Enter y to continue , n to stop')
    y=input('Do you want to continue ')
    if y=='y':
        continue
        print('-------------------------------')
    elif y=='n':
        print('-------------------------------')
        break
    else:
        print('-------------------------------')
        break
        
```
